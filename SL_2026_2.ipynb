```python
import GEOparse
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. Load Dataset
print("Loading GSE231713...")
gse = GEOparse.get_GEO(geo="GSE231713", destdir="./")

# 2. Advanced Extraction (Safely merging samples)
data_list = []
for gsm_name, gsm in gse.gsms.items():
    if not gsm.table.empty:
        # Standardizing column names to avoid KeyErrors
        temp_df = gsm.table.copy()
        # Use the first column as index and 'VALUE' as data
        id_col = temp_df.columns[0]
        temp_df = temp_df.set_index(id_col)[['VALUE']]
        temp_df.columns = [gsm_name]
        data_list.append(temp_df)

# Merge all valid samples into one DataFrame
data = pd.concat(data_list, axis=1)

# 3. Handle Metadata
metadata = gse.phenotype_data
labels = metadata['characteristics_ch1.0.condition']

# 4. Clean Data (Remove rows with NaN values)
data = data.dropna()

# 5. Gene Selection (Top 50 Variable Genes)
gene_variance = data.var(axis=1)
top_50_genes = gene_variance.sort_values(ascending=False).head(50).index
selected_data = data.loc[top_50_genes]

# 6. Final Summary
print("\n--- Process Completed Successfully ---")
print(f"Final Data Shape: {data.shape}")
print(f"Selected Matrix Shape: {selected_data.shape}")
print(f"Biological Groups: {list(labels.unique())}")
```
